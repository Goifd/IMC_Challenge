```python
import json

path = "D:/Projects/IMC_Challenge/logs/e2978d1b-fb60-4c9d-a49c-643b32a1e67a (3).log"

# Function to parse a single log entry
def parse_log_entry(entry):
    # Convert the log entry to a dictionary
    entry_dict = json.loads(entry)
    # Extract the lambda log and timestamp
    lambda_log = entry_dict['lambdaLog']
    timestamp = entry_dict['timestamp']
    # Further processing can be done here as needed
    return lambda_log, timestamp

# Read the log file
with open(path, 'r') as file:
    log_data = file.read()# read(skiprows=range(0, 1))

# Split the log data into individual log entries
log_entries = log_data.strip().split('\n}\n{\n')

# Process each log entry
for i, entry in enumerate(log_entries):
    entry = '{' + entry + '}'
    log_entries[i] = entry
    if i!=0 and i<1999:
        lambda_log, timestamp = parse_log_entry(entry)
        print(f'Timestamp: {timestamp}, \n Lambda Log: {lambda_log} \n')
    

```
