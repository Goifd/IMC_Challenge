```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

prices_path_0 = r'D:/Projects/IMC_Challenge/Round_3/data/prices_round_3_day_0.csv'
prices_path_1 = r'D:/Projects/IMC_Challenge/Round_3/data/prices_round_3_day_1.csv'
prices_path_2 = r'D:/Projects/IMC_Challenge/Round_3/data/prices_round_3_day_2.csv'

df_0 = pd.read_csv(prices_path_0, sep=";")
df_1 = pd.read_csv(prices_path_1, sep=";")
df_2 = pd.read_csv(prices_path_2, sep=";")



merged = pd.concat([df_0,df_1,df_2], axis=0)
merged = merged.fillna(0) 
merged = merged.drop(['day','timestamp'], axis=1)
merged['mid_price'] = (merged['ask_price_1']+merged['bid_price_1'])/2

df_bask = merged.loc[merged["product"] == "GIFT_BASKET"]
df_bask.index = np.arange(1,30001)

df_choc = merged.loc[merged["product"] == "CHOCOLATE"]
df_choc.index = np.arange(1,30001)
df_straw = merged.loc[merged["product"] == "STRAWBERRIES"]
df_straw.index = np.arange(1,30001)
df_rose = merged.loc[merged["product"] == "ROSES"]
df_rose.index = np.arange(1,30001)

df_bask['synthetic'] = df_rose['mid_price']+6*df_straw['mid_price']+4*df_choc['mid_price']
df_bask['diff'] = df_bask['mid_price']-df_bask['synthetic']


#merged.index = np.arange(1,30004)

# df_1['ORCHIDS_diff'] = df_1.ORCHIDS.diff(3)
# df_1['ORCHIDS_rolling10'] = df_1['ORCHIDS'].rolling(10).mean()
# df_1['ORCHIDS_rolling20'] = df_1['ORCHIDS'].rolling(20).mean()

# df_1['HUMIDITY_rolling20'] = df_1['HUMIDITY'].rolling(10).mean()
# df_1['HUMIDITY_rolling20_sq'] = df_1['HUMIDITY'].rolling(10).mean()**2
# df_1['HUMIDITY_rolling20_cb'] = df_1['HUMIDITY'].rolling(10).mean()**3
# df_1['HUMIDITY_rolling20_fth'] = df_1['HUMIDITY'].rolling(10).mean()**4

# df_1['SUNLIGHT_rolling20'] = df_1['SUNLIGHT'].rolling(10).mean()
# df_1['SUNLIGHT_rolling20_sq'] = df_1['SUNLIGHT'].rolling(10).mean()**2
# df_1['SUNLIGHT_rolling20_cb'] = df_1['SUNLIGHT'].rolling(10).mean()**3
# df_1['SUNLIGHT_rolling20_fth'] = df_1['SUNLIGHT'].rolling(10).mean()**4

# df_1['SUN_HUM']=df_1['HUMIDITY']*df_1['SUNLIGHT']


# print(merged.shape)
# print(merged.keys())
# print(merged.describe())
# print(merged.corr())
```

```python
plt.plot(df_bask['mid_price'],label='mid_price')
plt.plot(df_bask['synthetic'])
plt.legend()
print(df_bask['mid_price'].corr(df_bask['synthetic']))
```

```python
def trade_signal(row):
    """
    Function to generate trading signals based on price crossing moving average
    """
    if row['mid_price'] > row['Moving_Average_10']:
        return 'Buy'
    elif row['Price'] < row['Moving_Average_10']:
        return 'Sell'
    else:
        return 'Hold'
    
df_bask['Trade_Signal'] = df_bask.apply(trade_signal, axis=1)
plt.plot(df_bask['diff'].rolling(10).mean()[:100])
plt.plot(df_bask['diff'][:100],'.-')
```
